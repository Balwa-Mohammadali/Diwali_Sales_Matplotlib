```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv('Diwali Sales Data.csv',encoding='latin')

# print(df.head())

df=df.dropna(subset=['Gender','Age Group','Age','State','Occupation','Amount','Orders','Zone'])

plt.figure(figsize=(7,3))
gender=df['Gender'].value_counts()
plt.bar(gender.index,gender.values,color=['pink','yellow'],width=.5)
plt.title('gender ratio ')
plt.xlabel('gender')
plt.ylabel('no. of male/female')
plt.tight_layout()
plt.show()
plt.savefig('graph1.png',dpi=100,bbox_inches='tight')


plt.figure(figsize=(5,5))
age_group=df['Age Group'].value_counts()
plt.pie(age_group.values,labels=age_group.index,autopct='%1.1f%%',colors=['green','grey','lightblue','pink','yellow','red','orange'])
plt.title('pie chart of age group ')
plt.tight_layout()
plt.show()
plt.savefig('graph2.png',dpi=100,bbox_inches='tight')


plt.figure(figsize=(7,4))
age=df['Age']
plt.hist(age,bins=15,edgecolor='black',color='purple')
plt.title('Customer Age distribution')
plt.xlabel('Age')
plt.ylabel('no. of customer')
plt.grid()
plt.tight_layout()
plt.show()
plt.savefig('graph3.png',dpi=100,bbox_inches='tight')



plt.figure(figsize=(10,5))
state=df['State'].value_counts()
plt.scatter(state.index,state.values,color='red',label='state customber')
plt.title('Customer state distribution')
plt.xlabel('State')
plt.ylabel('no. of customer')
plt.legend()
plt.xticks(rotation=70)
plt.tight_layout()
plt.show()
plt.savefig('graph4.png',dpi=100,bbox_inches='tight')


plt.figure(figsize=(6,3))
zone=df['Zone'].value_counts()
plt.scatter(zone.index,zone.values,color='purple',marker='^',label='zone customber')
plt.title('Customer zone distribution')
plt.xlabel('zone')
plt.ylabel('no. of user')
plt.legend()
plt.tight_layout()
plt.show()
plt.savefig('graph5.png',dpi=100,bbox_inches='tight')


plt.figure(figsize=(8,4))
occ=df.groupby('Occupation')['Orders'].sum().sort_values(ascending=False)
plt.bar(occ.index,occ.values,color='darkviolet')
plt.title('order per occupation')
plt.xlabel('occupation')
plt.ylabel('no. of orders')
plt.xticks(rotation=70)
plt.tight_layout()
plt.show()
plt.savefig('graph6.png',dpi=100,bbox_inches='tight')


plt.figure(figsize=(8,4))
occ_amt=df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False)
plt.barh(occ_amt.index,occ_amt.values,color='purple')
plt.title('Amount spend per occupation')
plt.xlabel('Amount spends')
plt.ylabel('occupation')
plt.tight_layout()
plt.show()
plt.savefig('graph7.png',dpi=100,bbox_inches='tight')


plt.figure(figsize=(8,4))
age_amt=df.groupby('Age Group')['Amount'].mean()
plt.bar(age_amt.index,age_amt.values,color='darkblue',width=.5)
plt.title('avg. Amount spend by age group')
plt.xlabel('Age group')
plt.ylabel('Avg. amount')
plt.tight_layout()
plt.show()
plt.savefig('graph8.png',dpi=100,bbox_inches='tight')


age_spd_amt=df.groupby('Age')['Amount'].sum()
age_od=df.groupby('Age')['Orders'].sum()

fig ,ax=plt.subplots(1,2,figsize=(12,5))

ax[0].plot(age_spd_amt.index,age_spd_amt.values,color='red')
ax[0].set_title('Amount spend per age ')
ax[0].set_xlabel('Age')
ax[0].set_ylabel('amount spend')

ax[1].plot(age_od.index,age_od.values,color='black')
ax[1].set_title('no. of orders per age ')
ax[1].set_xlabel('Age')
ax[1].set_ylabel('no. of orders')

fig.suptitle('age comparision with order and amount spend ')
plt.tight_layout()
plt.show()
plt.savefig('graph9.png',dpi=100,bbox_inches='tight')


plt.figure(figsize=(8,4))
plt.hist(df['Amount'],bins=30,edgecolor='black',color='yellow')
plt.title('Amount distribution')
plt.xlabel('Amount')
plt.ylabel('frequency')
plt.grid()
plt.tight_layout()
plt.show()
plt.savefig('graph10.png',dpi=100,bbox_inches='tight')



plt.figure(figsize=(8,4))
plt.scatter(df['Age'],df['Amount'],color='green')
plt.title('Age vs Amount spend ')
plt.xlabel('Age')
plt.ylabel('Amount spend')
plt.grid()
plt.tight_layout()
plt.show()
plt.savefig('graph11.png',dpi=100,bbox_inches='tight')


plt.figure(figsize=(8,4))
plt.scatter(df['Orders'],df['Amount'],color='orange')
plt.title('order vs Amount spend ')
plt.xlabel('orders')
plt.ylabel('Amount spend')
plt.xticks([1,2,3,4])
plt.grid()
plt.tight_layout()
plt.show()
plt.savefig('graph12.png',dpi=100,bbox_inches='tight')


grouped=df.groupby(['Occupation','Gender'])['Orders'].sum().unstack()
x=np.arange(len(grouped.index))
width=0.35

plt.figure(figsize=(10,6))
plt.bar(x-width/2,grouped['M'],width,label='Male',color='blue')
plt.bar(x+width/2,grouped['F'],width,label='Female',color='coral')

plt.xlabel('occupation')
plt.ylabel('total orders')
plt.title('order by occupation and gender')
plt.xticks(x,grouped.index,rotation=45,ha='right')
plt.legend()
plt.tight_layout()
plt.show()
plt.savefig('graph13.png',dpi=100,bbox_inches='tight')

```
